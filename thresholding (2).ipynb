```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
import sys
#from skimage.color import rgb2gray


def threshold(img, threshold):
  
    flat = np.ndarray.flatten(img)

    for i in range(flat.shape[0]):

        if (flat[i] * 255) > threshold:
            flat[i] = 255
        else:
            flat[i] = 0

    thresh_img = np.reshape(flat, (img.shape[0], img.shape[1]))
  
    return thresh_img

if __name__ == '__main__':
   
    import cv2

    import glob


    path = glob.glob("blurredm\*.png")
    output_path='lung/thresholdm'
    c=0
    for file in path:

        img = cv2.imread(file)
       
        
        grayscale =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        smoothed_img = cv2.threshold(grayscale, 50, 255, cv2.THRESH_TOZERO)[1]
        cv2.imwrite(f'thresholdm/malignant{c}.png',smoothed_img)
        c+=1
        cv2.waitKey(0)

        cv2.destroyAllWindows()
```
