```python
import cv2 
import numpy as np
import glob

path = glob.glob("contourb\*.png")
output_path='lung/rotateb'
c=0

for file in path:

        img = cv2.imread(file)
       
        #smoothed_img = cv2.flip(img, 0)


       # img_flip_lr = cv2.flip(img, 1)
        img_rotate= cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
      #  cv2.imwrite(f'flipm/flipvm{c}.png',smoothed_img)
      #  cv2.imwrite(f'flipm/fliphm{c}.png',img_flip_lr)
        cv2.imwrite(f'rotateb/rotateb{c}.png',img_rotate)
        c+=1
        cv2.waitKey(0)

        cv2.destroyAllWindows()

```
